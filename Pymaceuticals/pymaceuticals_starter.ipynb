## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
single_dataset = pd.merge(mouse_metadata, study_results, how = "left", on=["Mouse ID"])
# Display the data table for preview
single_dataset.head()
```

```python
# Checking the number of mice.
num_mice = single_dataset["Mouse ID"].nunique()
print("There are " + str(num_mice) + " mice in the dataset.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = single_dataset[single_dataset.duplicated(["Mouse ID", "Timepoint"])]
duplicate_mice

```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_data = duplicate_mice["Mouse ID"].unique()
dup_data
print("The mouse with duplicate data is " + str(dup_data) + ".")
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = single_dataset.loc[single_dataset["Mouse ID"] != "g989"]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice = clean_df["Mouse ID"].nunique()
print("There are " + str(mice) + " in the clean DataFrame.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

grouped_reg = clean_df.groupby(["Drug Regimen"])
grouped_reg

tumor_vol = round(grouped_reg["Tumor Volume (mm3)"].count(), 2)
tumor_vol

drug_mean = round(grouped_reg["Tumor Volume (mm3)"].mean(), 2)
drug_mean

drug_median = round(grouped_reg["Tumor Volume (mm3)"].median(), 2)
drug_median

drug_variance = round(grouped_reg["Tumor Volume (mm3)"].var(), 2)
drug_variance

drug_standard_deviation = round(grouped_reg["Tumor Volume (mm3)"].std(), 2)
drug_standard_deviation

drug_standard_error_mean = round(grouped_reg["Tumor Volume (mm3)"].sem(), 2)
drug_standard_error_mean
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
summary_stats = pd.DataFrame({"Count" : tumor_vol,
                              "Mean": drug_mean, 
                              "Median": drug_median, 
                              "Variance": drug_variance, 
                              "Standard Deviation": drug_standard_deviation, 
                              "SEM": drug_standard_error_mean})
summary_stats
```
